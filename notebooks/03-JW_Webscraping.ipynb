```python
def get_lingala_text(Bible_dict = Bible_NT_dict, output_path = "../dataset/lingala/"):
    """
    This function will help to save in text form, depening to the book and verse argument 
    given in parameter

    
    The data is maintly taken from the website https://www.jw.org/
    """
    try :
        for book in Bible_dict:
            for chapter in range(1, Bible_dict[book][2]+1):
                r_ling = rq.get('https://www.jw.org/ln/Biblioteke/biblia/bi12/mikanda/'\
                            +str(book)+'/'+str(chapter)+'/')
                
                
                html_ling = r_ling.content
                soup_ling = BeautifulSoup(html_ling, 'html.parser') 
                s_ling = soup_ling.find_all("div", {"class": "textSizeIncrement"})

                if len(Bible_dict) == 39: 
                    key_caracter = "A"
                else :
                    key_caracter = "B"

                i = 0
                
                
                if not os.path.exists(output_path):
                    os.mkdir(output_path)
                    
                with open(output_path+key_caracter+str(Bible_dict[book][0]).zfill(2)+"_"+str(chapter).zfill(3)+"_verse_0.txt", "w") as text_file:
                    text_file.write(str(Bible_dict[book][1])+" "+str(chapter)+"\n")
                    
                content_loop = s_ling[0].text.split('\n')
                
#                 ipdb.set_trace()
                with open(output_path+key_caracter+str(Bible_dict[book][0]).zfill(2)+"_"+str(chapter).zfill(3)+"_verse_1.txt", "w") as text_file:
                    text = ' '.join(re.findall(r'(?:[^\W\d_]+\d|\d+[^\W\d_])[^\W_]*|[^\W\d_]+', content_loop[1]))
                    text_file.write(text.strip()+"\n")
                        
                for verse_line in content_loop[2:]:
                    if verse_line != '':
                        verse = verse_line.strip().split()[0]
#                         ipdb.set_trace()
                        if verse.isdigit():
                            verse_store = verse

                            with open(output_path+key_caracter+str(Bible_dict[book][0]).zfill(2)+"_"+str(chapter).zfill(3)+"_verse_"+str(verse)+".txt", "w") as text_file: # a+
#                                 text = re.sub(r"[+]", ' ', verse_line) 
#                                 # .strip() to delect any leading and trailing whitespace
#                                 text_file.write(text[2:].strip()+"\n") 
                                
    
                                text = ' '.join(re.findall(r'(?:[^\W\d_]+\d|\d+[^\W\d_])[^\W_]*|[^\W\d_]+', verse_line))
                                text_file.write(text.strip()+"\n")


                        else:
                            with open(output_path+key_caracter+str(Bible_dict[book][0]).zfill(2)+"_"+str(chapter).zfill(3)+"_verse_"+str(verse_store)+".txt", "a") as text_file: # a+
#                                 text = re.sub(r"[+]", ' ', verse_line)
#                                 # .strip() to delect any leading and trailing whitespace
#                                 text_file.write(text.strip()+"\n")
        
        
                                text = ' '.join(re.findall(r'(?:[^\W\d_]+\d|\d+[^\W\d_])[^\W_]*|[^\W\d_]+', verse_line))
                                # .strip() to delect any leading and trailing whitespace
                                text_file.write(text.strip()+"\n")
                    
            
    except Exception as inst:      
        traceback.print_exc()
```
