```python
# Pythono3 code to rename multiple 
# files in a directory or folder 

# importing os module 
import os 
from random import shuffle

skipped_files = set()

# Function to rename multiple files 
def TED_format(repository_1, repository_2, path_to_Save="../dataset", thres = 80):

    # given that the file in the two repository have the same name, 
    # we can only loop on one of them 
    
        
    if os.path.exists(f"{path_to_Save}/train.en-ln.csv"):
        os.remove(f"{path_to_Save}/train.en-ln.csv")
    
#     if os.path.exists("../dataset/valid.en-ln.csv"):
#         os.remove("../dataset/valid.en-ln.csv")
                  
    if os.path.exists(f"{path_to_Save}/test.en-ln.csv"):
        os.remove(f"{path_to_Save}/test.en-ln.csv")
                      
    # # This help to get the total number of files in arepository         
    # nb_file1 = len([name for name in os.listdir(repository_1) if os.path.isfile(os.path.join(repository_1, name))])
    # nb_file2 = len([name for name in os.listdir(repository_2) if os.path.isfile(os.path.join(repository_2, name))])
    

    # assert nb_file1 == nb_file2, f"Number of files mismatch between {repository_1} and {repository_2}" 
    
    # split = nb_file1*(thres/100)

    # # This need to be insert manually as colab return error getting it automatically 
    split = 29663*(thres/100)
    
    # This help suffle the files in the repository 
    list_of_files = os.listdir(repository_1)
    shuffle(list_of_files)
    
    for count, filename in enumerate(list_of_files): 
        
        name = filename.split(".")[0]
        
        language_1 = ""
        
        language_2 = ""
                
        # open the and read file on in the first repository             
        with open(repository_1+filename, 'r') as f:  # Open file for read
            for textline in f:           # Read line-by-line
                if language_1 == "":
                    language_1 += textline.split('\n')[0]
                else:
                    language_1 += " " + textline.split('\n')[0]
            
        # open the and read file on in the second repository    
        with open(repository_2+filename, 'r') as f:  # Open file for read
            for textline in f:           # Read line-by-line
                if language_2 == "":
                    language_2 += textline.split('\n')[0]
                else:
                    language_2 += " " + textline.split('\n')[0]

        # This help ovoid issue caused by files with empty content
        if name.strip() =="" or language_1.strip() == "" or language_2.strip() == "":
            skipped_files.add(name)
            continue
            
        
        if count <= split:
            with open(f"{path_to_Save}/train.en-ln.csv", "a+", encoding="utf-8") as text_file:
                text_file.write(name+".wav\t"+language_1+"\t"+language_2)
                text_file.write("\n")
        else:
            with open(f"{path_to_Save}/test.en-ln.csv", "a+", encoding="utf-8") as text_file:
                text_file.write(name+".wav\t"+language_1+"\t"+language_2)
                text_file.write("\n")
                  
#         elif count <=7800:
#             with open("../dataset/valid.en-ln.csv", "a+", encoding="utf-8") as text_file:
#                 text_file.write(name+".wav\t"+language_1+"\t"+language_2)
#                 text_file.write("\n")          
  
            

# # Driver Code 
# if __name__ == '__main__': 
#     # Calling main() function 
#     # main() 
    
```
