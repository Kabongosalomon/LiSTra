```python
# Pythono3 code to rename multiple 
# files in a directory or folder 

# importing os module 
import os 

# Function to rename multiple files 
def generatorFunction(file_path, output_path="../dataset",
                      lang_1="en", lang_2="fr",separator="\n"):

    if os.path.exists(f"{output_path}/{lang_1}.vocab"):
        os.remove(f"{output_path}/{lang_1}.vocab")

    if os.path.exists(f"{output_path}/{lang_2}.vocab"):
        os.remove(f"{output_path}/{lang_2}.vocab")

    if os.path.exists(f"{output_path}/{lang_1}-{lang_2}.vocab"):
        os.remove(f"{output_path}/{lang_1}-{lang_2}.vocab")
    
    lang1_set = set()
    lang2_set = set()
    lang1_lang2_set = set()
    
                
    with open(file_path, 'r') as f:  # Open file for read
        for textline in f:           # Read line-by-line
            _, line_lang_1, line_lang_2 = textline.split(separator)
                
            for word in line_lang_1.split():
                with open(f"{output_path}/{lang_1}.vocab", "a+", encoding="utf-8") as text_file:
                    if word not in lang1_set:
                        text_file.write(word)
                        text_file.write(" ")
                        lang1_set.add(word)
            
            for word in line_lang_2.split():
                with open(f"{output_path}/{lang_2}.vocab", "a+", encoding="utf-8") as text_file:
                    if word not in lang2_set:
                        text_file.write(word)
                        text_file.write(" ")
                        lang2_set.add(word)
                    
            for word in (line_lang_1+" "+line_lang_2).split():
                with open(f"{output_path}/{lang_1}-{lang_2}.vocab", "a+", encoding="utf-8") as text_file:
                    if word not in lang1_lang2_set:
                        text_file.write(word)
                        text_file.write(" ")
                        lang1_lang2_set.add(word)
```

```python
# Driver Code 
if __name__ == '__main__': 
    # Calling main() function 
    # main() 
    generatorFunction(file_path="../dataset/train.en-ln.csv", output_path="../dataset",
                      lang_1="en", lang_2="ln", separator="\t")
```
